```python
import numpy as np
import matplotlib.pyplot as plt
from utils.utils import read_Expre_tsv
```

```python
total_expressions = read_Expre_tsv("data/expression_cov_1024_200_bulk.tsv")
```

```python
# The mean of total expressions
mean = np.mean(total_expressions, axis=1)
print(f'Max value of the mean gene expression is {mean.max()}')
print(f'Min value of the mean gene expression is {mean.min()}')

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(mean, bins='auto')
plt.title('Full Data')
plt.xlabel('Mean Gene Expressions')
plt.ylabel('Sequence Number')

# trim the data
plt.subplot(1, 2, 2)
trimmed_mean = mean[(mean > np.percentile(mean, 5)) & (mean < np.percentile(mean, 95))]
plt.hist(trimmed_mean, bins='auto')
plt.title('Trimmed Data')
plt.xlabel('Mean Gene Expressions')
plt.ylabel('Sequence Number')

plt.tight_layout()

plt.show()
```
